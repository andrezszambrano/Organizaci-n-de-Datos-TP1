```python
import seaborn as sns
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import warnings
warnings.filterwarnings('ignore')
dfFeat = pd.read_csv('./Visualizaciones/dengue_features_train.csv')
dfLabels = pd.read_csv('./Visualizaciones/dengue_labels_train.csv')
```

```python
dfFeat
```

```python
dfLabels
```

```python
df = pd.merge(dfFeat, dfLabels, on = ['city', 'year', 'weekofyear'])
```

```python
df
```

```python
dfSj = df.loc[df['city'] == 'sj', :]
dfIq = df.loc[df['city'] == 'iq', :]
```

```python
dfSj.loc[dfSj['weekofyear'] <= 13, 'Quarter'] = 'Q1'
dfSj.loc[(13 < dfSj['weekofyear']) & (dfSj['weekofyear'] <= 26), 'Quarter'] = 'Q2'
dfSj.loc[(26 < dfSj['weekofyear']) & (dfSj['weekofyear'] <= 39), 'Quarter'] = 'Q3'
dfSj.loc[(39 < dfSj['weekofyear']) & (dfSj['weekofyear'] <= 52), 'Quarter'] = 'Q4'
dfSj.head(20)
```

```python
dfSj.loc[dfSj['total_cases'] <= 50, 'cant_casos'] = 'entre0y50'
dfSj.loc[(50 < dfSj['total_cases']) & (dfSj['total_cases'] <= 100), 'cant_casos'] = 'entre50y100'
dfSj.loc[(100 < dfSj['total_cases']) & (dfSj['total_cases'] <= 150), 'cant_casos'] = 'entre100y150'
dfSj.head(20)
```

```python
dfSj['cant_casos'].value_counts()
```

```python
dfSj
```

```python
dfSj = dfSj.loc[dfSj['total_cases'] < 150, :]
dfSj
```

```python
dfSj.loc[dfSj['precipitation_amt_mm'] <= 50, 'cant_precipitacion'] = 'poca'
dfSj.loc[(50 < dfSj['precipitation_amt_mm']) & (dfSj['precipitation_amt_mm'] <= 100), 'cant_precipitacion'] = 'media'
dfSj.loc[(100 < dfSj['precipitation_amt_mm']), 'cant_precipitacion'] = 'mucha'
dfSj.head()
```

```python
dfAux = dfSj.loc[dfSj['cant_casos'] != 'entre100y150']
```

```python
fig_dims = (10, 10)
fig, ax = plt.subplots(figsize = fig_dims)
sns.boxplot(data = dfAux, y = 'total_cases', x = 'Quarter', hue = 'cant_precipitacion',
            order = ['Q1', 'Q2', 'Q3', 'Q4'], hue_order = ['poca', 'media', 'mucha'],
            color = 'blue', ax = ax, linewidth = 3, whis = 2, fliersize = 2).set(title = 'Densidad de casos totales de dengue en San Juan por cuatrimestre, según precipitación')
plt.xlabel("Número de cuatrimestre")
plt.ylabel("Casos totales por semana")
plt.savefig('visu3CasosDengueEnSanJuanPorCuatrimestre.jpg',
           format='jpeg', dpi=200, bbox_inches = "tight")
```

```python
dfIq.loc[dfIq['weekofyear'] <= 26, 'Semester'] = 'S1'
dfIq.loc[(26 < dfIq['weekofyear']), 'Semester'] = 'S2'
dfIq.head()
```

```python
dfIq.loc[dfIq['total_cases'] <= 50, 'cant_casos'] = 'entre0y50'
dfIq.loc[(50 < dfIq['total_cases']) & (dfIq['total_cases'] <= 100), 'cant_casos'] = 'entre50y100'
dfIq.loc[(100 < dfIq['total_cases']) & (dfIq['total_cases'] <= 150), 'cant_casos'] = 'entre100y150'
dfIq.head()
```

```python
dfIq["cant_casos"].value_counts()
```

```python
dfIq = dfIq.loc[dfIq["cant_casos"] == 'entre0y50']
dfIq = dfIq.loc[(dfIq["year"] != 2001) & (dfIq["year"] != 2000)]
```

```python
fig_dims = (12, 8)
fig, ax = plt.subplots(figsize = fig_dims)
sns.violinplot(data = dfIq, x = 'year', y = 'total_cases', hue = 'Semester', color = 'green',
               orientation = 'horizontal', ax = ax, cut = 0, linewidth = 1.5, split = True).set(title = 'Densidad de casos totales de dengue en Iquitos por año, según semestre')
plt.xlabel("Año")
plt.ylabel("Casos totales por semana")
plt.savefig('visu4CasosDengueEnIquitosPorAños.jpg',
           format='jpeg', dpi=200, bbox_inches = "tight")
```
