# Ejercicio 49
Si decimos que un usuario sabe un idioma cuando tiene un nivel de babel mayor o
igual a 1, para aquellos que editaron una de las versiones actuales del contenido,
¿Cuál es la tasa de revisiones sin comentario que realizan en función de los idiomas
que saben?

```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
dfLanguages = pd.read_csv('./WikipediaDataset/languages.csv')
dfContents = pd.read_csv('./WikipediaDataset/contents.csv', usecols = ["revisor_id", "revisor_comment"])
```

```python
dfLanguages 
```

```python
dfLanguages.info()
```

```python
dfLanguages = dfLanguages[dfLanguages['babel_lang'].notna()]
dfLanguages["babel_level"] = dfLanguages["babel_level"].replace("N", 4.5)
```

```python
dfLanguages.info()
```

```python
dfLanguages["babel_level"] = dfLanguages["babel_level"].astype(str).astype(float)
dfLanguages.info()
```

```python
dfLanguages = dfLanguages.loc[dfLanguages["babel_level"] >= 1].reset_index(drop = True)
dfLanguages 
```

```python
dfLanguages["cant_idiomas"] = dfLanguages.groupby("babel_user")["babel_lang"].transform("count")
dfLanguages
```

```python
dfLanguages.drop_duplicates(subset = ["babel_user"], inplace = True, ignore_index = True)
```

```python
dfLanguages.rename(columns = {'babel_user': 'id_usuario'}, inplace = True)
dfLanguages.head()
```

```python
dfLanguages = dfLanguages.loc[:, ["id_usuario", "cant_idiomas"]]
dfLanguages
```

```python
dfContents
```

```python
dfContents = dfContents[dfContents['revisor_id'].notna()]
```

```python
dfContents.rename(columns = {'revisor_id': 'id_usuario'}, inplace = True)
dfContents.head()
```

```python
dfMerge = pd.merge(dfLanguages, dfContents, how = "inner", on = ["id_usuario"])
dfMerge
```

```python
dfMerge.groupby("id_usuario")["revisor_comment"].count()
```

```python
dfMerge.groupby("id_usuario")["revisor_comment"].count()
```

```python
dfMerge["revisiones_con_comentario"] = dfMerge.groupby("id_usuario")["revisor_comment"].transform("count")
dfMerge["revisiones_totales"] = dfMerge.groupby("id_usuario")["revisor_comment"].transform("size")
dfMerge.drop_duplicates(subset = ["id_usuario"], inplace = True, ignore_index = True)
dfMerge["tasa_revisiones_sin_comentario/revisiones_totales"] = (dfMerge["revisiones_totales"] - dfMerge["revisiones_con_comentario"])/dfMerge["revisiones_totales"] 
dfMerge
```

```python
dfMerge = dfMerge.loc[:, ["id_usuario", "cant_idiomas", "tasa_revisiones_sin_comentario/revisiones_totales"]]
dfMerge
```
